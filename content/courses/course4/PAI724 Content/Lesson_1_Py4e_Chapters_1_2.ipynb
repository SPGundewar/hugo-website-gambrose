We can also assign data to variables or place holders. 

```python
w = 8					
print(w) 			

	# placeholders can be anything, but it is best practice to keep them descriptive 
```

```python
# We can also use code to manipulate these variables. 
w = 8					
print(w * 14)


# while we have used the parameter 'w' in the expression, the value of 'w' stays the same and we haven't 'saved' or expression
```

```python
# We can overwrite the variable 'w' by assigning the expression again to 'w'
w = 8					
print(w)
w = w * 14				
print(w) 	

# In this simple code we also make a fundamental point: logical chain
```

```python
# But it is best practice to save our expression by creating a new parameter 'y' incase we want to call 'w' in the future 
w = 8					
print(w)
y = w * 14				
print(y) 	

# In this simple code we also make a fundamental point: logical chain
```

Say we want to report the height of our respondent in inches.

```python
height = input("How tall are you (round to the nearest foot)?")
print("you are about " + height +" feet tall")
h_inch = height * 12
print("you are about " ,  h_inch ," inches tall")
```

Why we are getting 666666666666 ?		

While height is referencing the input variable that variable is a string variable. Python is manipulating what you are giving it. It has multiplied the string ‘6’ twelve times. 

Python sees all inputted data as string data. 


```python
height = input("How tall are you (round to the nearest foot)?")
print("you are about " , height ," feet tall")
h = int(height)                                         #Must identify the data as an integer
h_inch = height * 12
print("you are about " , h_inch ," inches tall")
```

Still a problem.

Violation of the logical chain. We change height to an integer and assign it to h, but we do not use h in the next line of code.

```python
height = input("How tall are you (round to the nearest foot)?")
print("you are about " , height ," feet tall")
h = int(height)
h_inch = h * 12                                          # maintain the logical chain
print("you are about " , h_inch ," inches tall")
```
