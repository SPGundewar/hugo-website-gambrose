Furthermore, we might want to count things that are more interesting, such as the number of participants in this dataset. If we go back to reading the file.

We can see that each line that identifies a new participant starts with “ID#:”
So we can write a for-loop that only counts lines that start with “ID#:”


```python
count = 0
with open("/content/E5WP1_Data_File.txt") as text:
  for line in text:
    if line.startswith("ID#"):
        count = count + 1
print(count)
```

Additionally, we can use the endswith command to count the lines of text. While it does not show up in the panel, text files have 'hidden commands'. One of these commands is '\n' command or the 'new line' command. This is the code that actually tells the file to return to a new line. Since we know this, we can leverage it to count the lines in the text file.

```python
count = 0
with open("/content/E5WP1_Data_File.txt") as text:
  for line in text:
    if line.endswith("\n"):
        count = count + 1
print(count)
```

We see that our count here is one different than our count before. Why do you think that is?

The last time does not have a '\n' because it does not return to a new line, but the file simply ends. We must account for this.


```python
count = 0
with open("/content/E5WP1_Data_File.txt") as text:
  for line in text:
    if line.endswith("\n"):
        count = count + 1
print(count+1)
```

# List Operations

Let us switch gear a bit to think about list operations.

Let’s start out with a list of numbers


```python
numbers = [4,2,8,4,5]
```

Lists, unlike strings are mutable. We can change them directly.

For instance, in the spirit of last week's typo, let's say we meant to type 9 instead of 8 that is currently in our list.

```python
numbers = [4,2,8,4,5]
numbers[2] = 9
print(numbers)
```

Say we want to know the square root of all of the numbers in this list



```python
numbers = [4,2,8,4,5]
import math
for i in range(len(numbers)):
    numbers[i] = math.sqrt(numbers[i])
print(numbers)
```

What is the difference between this for loop and the ones that we used two weeks ago?

While previously we used a list of numbers to generate a single outcome, here we are replacing the iteration variable back into its position in the list after we complete some computation. Here we take four from the 0 position in the list, we calculate it's square root, and place 2 back into position 0. Next, we move to position 1 and do the same thing. 

So let's think through this a bit more

We also are able to manipulate the dataset or merge multiple datasets. Say we want to remove 8 from the list. We can use pop and its position to remove it.

```python
numbers = [4,2,8,4,5]
new_numbers = numbers.pop(2)
print(new_numbers)
print(numbers)
import math
for i in range(len(numbers)):
    numbers[i] = math.sqrt(numbers[i])
print(numbers)
```

If we do not know the position of 8 but still want to remove it, we can use remove and call the object in the string directly

```python
numbers = [4,2,8,4,5]
new_numbers = numbers.remove(8)
print(new_numbers)
print(numbers)
import math
for i in range(len(numbers)):
    numbers[i] = math.sqrt(numbers[i])
print(numbers)

```

As you can see, new_numbers now holds the ‘None’ value. This has to do with the functionality of the command. While pop called a position and identified the value in the position, remove removes the value but does not report the position or value.

Finally, If our list of values is split over two variables, we can merge using extend.


```python
numbers = [4,2,8,4,5]
numbers2 = [3,6,1]
new_numbers = numbers.extend(numbers2)
print(new_numbers)
print(numbers)
import math
for i in range(len(numbers)):
    numbers[i] = math.sqrt(numbers[i])
print(numbers)
```

Again, we can see, new_numbers now holds the ‘None’ value.

It is important to think through what each command is doing for us and how it should be leveraged. 



Additionally, we can see that lists are mutable. We can change the specific positions in the list that we desire to change. This is different than the slicing that we need to do last week with the immutable strings from last week.
