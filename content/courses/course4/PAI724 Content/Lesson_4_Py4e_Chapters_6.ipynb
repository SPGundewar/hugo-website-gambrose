Let us start off with an important point – Strings are immutable.

What does this mean?

In practice this means that we are not changing the string directly, but we have to call up specific potions of a string and concertante with a new string.

Mechanically, let's try to think through this as trying to fix an incorrect ordering of numbers.


```python
hi = "Hello, 0AI 724"
print(hi)
```

Obviously, we want PAI instead of OAI.

But first, let's think about position and length in string variables.



```python
hi = "Hello,x0AIx724"
print(hi)
length = len(hi)
print(length)
```

Let's print the last letter in our string, using the length we just calculated.

```python
print(hi[length]) 
```

Why do we get the error 'string out of range'?

Python starts with 0 when counting position.

```python
print(hi[0])
print(hi[length-1])
```

Given the numbering procedure in Python, the typo is in the 7th position of the string... Count out ... So if we want to call up the typo we can call it with the square brackets following the variable name of out string.


```python
hi = "Hello,x0AIx724"
print(hi[7])
```

Now if strings were mutable, if we could change them directly, we could reassign the correct letter to that position.

```python
hi = "Hello,x0AIx724"
hi[7] = "P"
print(hi)
```

But the error tells us that 'str' object does not support item assignment—in other words strings are immutable and we have to cut up out typo string and paste in the correct character.



```python
hi = "Hello,x0AIx724"
hi2 = hi - "0AI 724"
print(hi2)
```

This runs into a similar immutability problem. Despite specifying the specific characters of the string we want to have ‘subtracted off’, python only knows how to interact with the string given first identifying the string and then identifying positions in the string.


```python
hi = "Hello,x0AIx724"
hi2 = hi[0:7]
print(hi2)
```

So we have identified our string; we identified the first position to be included; and the first position to be excluded

```python
hi = "Hello,x0AIx724"
hi2 = hi[8:13]
print(hi2)
```

To get the last character we can just leave the final position open—meaning until the end of the string.

```python
hi = "Hello,x0AIx724"
hi2 = hi[7:]
print(hi2)
```

In the same way, leaving the space open means from the beginning.

```python
hi = "Hello,x0AIx724"
hi2 = hi[:7]
print(hi2)

```

And now again, we have all of our correct text up to our typo.

```python
hi = "Hello,x0AIx724"
hi2 = hi[:7]
hi3 = hi[]        # Now if I want all of the characters after the typo, 
print(hi2)        # what should I put in the brackets?
print(hi3)

```

```python
hi = "Hello,x0AIx724"
hi2 = hi[:7]
hi3 = hi[8:]
print(hi2)
print(hi3)
```

So now as we can see, we have out separate chunks of the original hi string that be concatenated back together without the typo.


```python
hi = "Hello,x0AIx724"
hi2 = hi[:7]
hi3 = hi[8:]
print(hi2)
print(hi3)
hi4 = hi2 + "P" + hi3    # Here, remember that ‘+’ is a concatenation 
print(hi4)               # function for strings.
```

Okay, this is fine and all, but this whole process is a little tedious, and we just learned how for-loops can count for us! Let’s try to leverage that and have Python take a little bit of workload off of our shoulders.

Let us start by creating a count for-loop that will find the position of the typo.



```python
hi = "Hello,x0AIx724"
before = 0
for letter in hi:
    if letter == "0":
        break
    else:
        before = before + 1
print(before)
```

Now that we have the position and assigned that position to the variable ‘count’ we can call our position to get the first part of our string.

```python
hi = "Hello,x0AIx724"
before = 0
for letter in hi:
    if letter == "0":
        break
    else:
        before = before + 1
print(before)
print(hi[:before])
```

Remember before is just a place holder for the value 7. So we can add 1 to before to skip over the typo.

```python
hi = "Hello,x0AIx724"
before = 0
for letter in hi:
    if letter == "0":
        break
    if letter > " ":
        before = before + 1
print(before)
print(hi[:before])
after = before + 1
print(hi[:before]+"P"+hi[after:])
```

Or, to be efficient coders, we can place the calculation right in the brackets

```python
hi = "Hello,x0AIx724"
before = 0
for letter in hi:
    if letter == "0":
        break
    if letter > " ":
        before = before + 1
print(before)
print(hi[:before])
print(hi[:before]+"P"+hi[before +1:])
```
