```python
string = "Hello PAI 724! How are we doing today?"
d = dict()
for c in string:
    if c not in d: 		                                # What is functionally happening in the first conditional statement?
        d[c]=1		                                    # if the letter is not in the dictionary, we are placing both the key and value.
    else: 			                                      # What is functionally happening in the first conditional statement?
        d[c] = d[c]+1	                                # if the letter is in the dictionary, we are adding one to the value of the 
print(d)				                                      # associated key.	
```

Now, lets say I just want the names and do not want the @syr.edu


```python
emails = ["nick@syr.edu", "anna@syr.edu", "nick@syr.edu", "nick@syr.edu", "jiho@syr.edu"]
names = dict()
syr_edu = "@syr.edu"
for email in emails:
    email = email.translate(email.maketrans('','',syr_edu))
    if email not in names:		                                                              # for maketrans() the first position is what characters are being 
        names[email] = 1			                                                              # replaced, the second is what characters will be doing the 
    else:					                                                                          # replacing, and the last are characters that will be deleted.
        names[email] += 1			                                                              # In this case we are only interested in deleting characters.
print(names)
```

We can change the keys in the dictionary to get different counts


```python
emails = ["nick@syr.edu", "anna@syr.edu", "nick@syr.edu", "nick@syr.edu", "jiho@syr.edu"]
names = {"anna":0, "jiho": 0}
syr_edu = "@syr.edu"
for email in emails:
    email = email.translate(email.maketrans('','',syr_edu))
    if email in names:		 
        names[email] += 1			 
print(names)	

```
