First, let's start out by creating a basic while function.


````python
number = 1

while number < 10:
  print(number)
  number = number + 1

print('You have made it to 10')
````

In this iteration, we start with the value of number as 1, print the current value, and added 1 through each iteration. When the conditional execution is false, the *while* function leaves the loop and prints the final string.

What should I do if I want to print '10' before leaving the loop?

We can change the conditional statement:

````python
number = 1

while number <= 10:
  print(number)
  number = number + 1

print('You have made it to 10')
````

Additionally, we can write loops that are Semantically false:

````python
number = 1

while number <= 10:
  print(number)
  number = number - 1  # change from '+' to '-'

print('You have made it to 10')
````

In this case, we have created an infinite loop. Given the function in the loop:
```
  number = number - 1
```
...the evaluation will always be true; therefore the iteration will run infinitely. 

Thankfully we can stop the function by pressing the 'stop' button or the 'square' button that replaces the 'play' button.

To calculate mean we also need to know how many values are in our list, so we must add a count to our for loop.

````python
height = [5.4,7,6.2,5.9,6,5]
total = 0
count = 0
for itervar in height:			
	total = total + itervar	
	count = count + 1	#again, we have count on both the left and right of the equal sign
print(total)
print(count)
````

Now, for each iteration we go through the loop, we are also adding 1 to the variable count. 

Finally, we will calculate the mean.

````python
height = [5.4,7,6.2,5.9,6,5]
total = 0
count = 0
for itervar in height:			
	total = total + itervar	
	count = count + 1	
	mean = (total/count)
print(total)
print(count)
print(mean)
````

Why do I not need to create a variable before the iteration function as I did for total and count?

````python
height = [5.4,7,6.2,5.9,6,5]
total = 0
count = 0
for itervar in height:
  total = total + itervar	
  count = count + 1	
  mean = (total/count)
  print(' ')
  print(total)
  print(count)
  print(mean)
````

While we can make mathematical calculations using iterations, we can also place conditional functions in for loops.
