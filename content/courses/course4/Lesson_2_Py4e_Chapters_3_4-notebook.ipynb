This lesson will work through Conditional Execution using Boolean expressions and conditional operators as well as programing new functions in Python.


First, let's use and think through Boolean expressions. 


```python
x = 5
x == 5

# The double equal is a Boolean operator checking if the first operand is the 
# same as the second.
# More specifically, because we assigned 5 to the variable x in the first line,
# we evaluate 5 == 5 in the second line
```

Next, we can use the Boolean evaluation with the *if* command.


```python
x = 5
if x == 5:                  
  print("the variables are the same")
```

In the same way we can replace the integer 5 with the string 'five' 

```python
x = 'five'
if x == 'five':
	print("the variables are the same")

# the evaluation in the second line is still TRUE
```

To take this point one step further--

```python
x = 'five'
if x == 'five ':  # we can add a space after 'five' --> 'five '
	print("the variables are the same")


# While our understanding of the operand has not changed, the values of the two 
# operands are now different. The additional space in a strict Boolean sense 
# makes this statement false.
```

In the last version of the function, we receive no output. When looking at the second line of code, we have only given the function an expression to execute if the Boolean evaluation is TRUE. Because it is FLASE, there is no expression to execute.

We can add code to tell the function what to do in the case of a FLASE evaluation. We can use *else* to give instructions for these FALSE cases.

```python
x = 'five'
if x == 'five ': 
	print("the variables are the same")
else:    
  print("the variables are NOT the same")

# else is used to capture ALL OUTCOMES where the if evaluation is not TRUE
```

Now that we have the basics, lets make this a bit more interesting. If we begin with our in class example from last week. Say I am interested in knowing if people are taller or shorter than my own height.

```python
height = input("How tall are you (round to the nearest foot)?")
print("you are about", height, "feet tall")
if height > 6:
  print("taller")
else:
  print("shorter")
```

```python
height = input("How tall are you?")
print("you are", height, "feet tall")
h = float(height)         # We have to identify the input as a float or integer
if h > 6:
    print("you are taller")
else:
    print("you are shorter")
```

Test the above function. Choose a value above and below 6.

Now input 6. Here we have a *semantic error*. The syntax and logic sound, but *our instructions* yield a logical fallacy. 6 is not shorter than 6. Yet, in our code, all values not greater than 6 are identified as shorter than 6.

We can account for additional evaluation with *elif*. 


```python
height = input("How tall are you (round to the nearest foot)?")
print("you are about", height, "feet tall")
h = float(height)
if h > 6:
    print("you are taller")
elif h == 6:
    print("you are the same height")
else:
    print("you are shorter")

# elif is a contraction of the else and the if commands
# In this way, it recieves all of the FALSE evaluation from the if statement and
# executes another evaluation
```

Finally, if we are interested in having a range or values for *elif* instead of on specific term, we can use a logical *and*:

```python
height = input("How tall are you (round to the nearest foot)?")
print("you are about", height, "feet tall")
h = float(height)
if h > 6:
    print("you are taller")
elif h <= 6 and h >= 5:     # note <= and >= ; the order is important and the  
    print("you are about the same height")       # same as how you would say it
else:
    print("you are shorter")
```

#New Functions

In python there are two types of functions. *Void Functions* which will give you the same output every time and *Fruitful Functions*, which will interact with a variable give the function you construct. They have the same structure but work in different ways.

First, let’s start with void functions.

*def* is the python command that defines our new function. What comes after is the name of the new  function.
In the second line, we define the execution of the new function when it is called.

```python
def print_HI():
  print('Hello, PAI 724') 
```

*Are there any thoughts on why we don’t have any output?*

While we defined the new function, we never called the function to execute. 

```python
def print_HI():
  print('Hello, PAI 724') 

print_HI()
```

We can also construct a function that is multiple lines:


```python
def print_HI():
    print('Hello')
    print('PAI 724')
print_HI()
```

We can also layer functions to call it multiple times:

```python
def print_HI():
    print('Hello')
    print('PAI 724')
def print_HIs():
    print_HI()
    print_HI()
print_HIs()    #Another example of logical chain --> work backwards
```

Remember these are *void functions* because they give us the same response whenever we call them. If we want to interact with our new functions we need to use a *fruitful function*.

Using the same commands and structure we can change our void function into a fruitful function. We need to replace the parameters of out definition with variable placeholders.

```python
def names(X):  # the variable 'X' is out place holder
    print(X)
    print(X)
names("GRAHAM") #"GRAHAM" is assigned to the function. 
                   # functionally: X = "GRAHAM"

#Another example of logical chain --> work backwards
```

The input can be changes; hence, the function is fruitful.

```python
def names(X): 
    print(X)
    print(X)
names(7) 
```

While, at this point, this seems trivial, it can become important if we want to execute the same function given different variable inputs.

For instance, we can evaluate the mean of a set of numbers.

```python
def mean_val(a, b):
    add = a + b
    m = (add/2)
    print(m)

mean_val(5, 13) 
```

```python
def mean_val(a, b):
    add = a + b
    m = (add/2)
    print(m)

mean_val(5, 13, 2, 42, 1, 6, 17, 29, 31) 
```

While we have added inputs, we have not added placeholders in the definition to deal with these new inputs.

```python
def mean_val(a, b, c, d, e, f, g, h, i):
    add = a + b + c + d + e + f + g + h + i
    m = (add/9)
    print(m)

mean_val(5, 13, 2, 42, 1, 6, 17, 29, 31) 
```

This shows that fruitful functions are more flexible than void functions, but they are also very structured to the construction of the definition.  We will be learning more flexible application of codes next week when we learn *for loops*.
