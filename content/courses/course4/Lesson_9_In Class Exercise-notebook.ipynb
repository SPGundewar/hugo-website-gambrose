```r
library(tidyverse)
```

```r
data2 <- read_csv("/content/Lesson_10_EJ_Data.csv")
head(data2)
```

```r
 data3 <- data2 %>%
  group_by(Year, Part_Sec, Part_Position)%>%
  summarise(t_statm = n()) 

head(data3)
```

```r
 data4 <- data3 %>%
  ungroup() %>%
  group_by(Year, Part_Sec) %>%
  summarise(sd_statm = sd(t_statm), mean_statm = mean(t_statm))

head(data4)
```

```r
ggplot(data = data4, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1)
```

```r
ggplot(data = data4, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1) +
  geom_errorbar(aes(ymin=(mean_statm-sd_statm), ymax=(mean_statm+sd_statm),color = Part_Sec), width=.2, position = position_dodge(width = 0.2)) 
```

```r
ggplot(data = data4, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1) +
  geom_errorbar(aes(ymin=(mean_statm-sd_statm), ymax=(mean_statm+sd_statm),color = Part_Sec), width=.2, position = position_dodge(width = 0.2)) + 
  scale_color_manual(values=c("#999999", "red")) 
```

```r
ggplot(data = data4, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1) +
  geom_errorbar(aes(ymin=(mean_statm-sd_statm), ymax=(mean_statm+sd_statm),color = Part_Sec), width=.2, position = position_dodge(width = 0.2)) + 
  scale_color_manual(values=c("#999999", "red")) +
  scale_x_continuous(n.breaks=11)
```

```r
ggplot(data = data4, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1) +
  geom_errorbar(aes(ymin=(mean_statm-sd_statm), ymax=(mean_statm+sd_statm),color = Part_Sec), width=.2, position = position_dodge(width = 0.2)) + 
  scale_color_manual(values=c("#999999", "red")) +
  scale_x_continuous(n.breaks=11) + 
    labs(
      title = "Statements made by each sector across time",
      x = "Years",
      y = "Number of Statemetns per Year")+ 
    theme(plot.title = element_text(face = "bold", size =14, hjust = 0.5))
```

```r
 data5 <- data2 %>%
  group_by(Year, Part_Sec, Part_Position, types) %>%
  summarise(t_statm = n())

head(data5)
```

```r
 data6 <- data5 %>%
  ungroup() %>%
  group_by(Year, Part_Sec, types) %>%
  summarise(sd_statm = sd(t_statm), mean_statm = mean(t_statm))

head(data6)
```

```r
ggplot(data = data6, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1) +
  geom_point(aes(color = Part_Sec)) + 
  geom_errorbar(aes(ymin=(mean_statm-sd_statm), ymax=(mean_statm+sd_statm),color = Part_Sec), width=.1) + 
  scale_color_manual(values=c("#999999", "red"))+
  scale_x_continuous(n.breaks=11)+ 
    labs(
      title = "Types of statements made by each sector across time",
      x = "Years",
      y = "Number of Statemetns per Year")+ 
    theme(plot.title = element_text(face = "bold", size =14, hjust = 0.5))
```

```r
ggplot(data = data6, mapping = aes(x = Year, y = mean_statm, group = Part_Sec)) +
  geom_line(aes(color = Part_Sec, linetype = Part_Sec), size=1) +
  geom_point(aes(color = Part_Sec)) + 
  geom_errorbar(aes(ymin=(mean_statm-sd_statm), ymax=(mean_statm+sd_statm),color = Part_Sec), width=.1) + 
  scale_color_manual(values=c("#999999", "red"))+
  scale_x_continuous(n.breaks=11)+ 
  facet_wrap( ~ types, ncol=1)+ ### make a graph for each type of statement
    labs(
      title = "Types of statements made by each sector across time",
      x = "Years",
      y = "Number of Statemetns per Year")+ 
    theme(plot.title = element_text(face = "bold", size =14, hjust = 0.5))
```
